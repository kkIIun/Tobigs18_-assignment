```python
# 필요한 모듈 불러오기
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data
- 보시면 아시다시피 spam.csv는 라벨이 있는 데이터입니다. 물론 7주차 주제가 텍스트 기초인만큼 텍스트만 활용하셔도 되고, 라벨까지 활용하셔서 모델을 돌려보셔도 좋습니다 :)

```python
spam = pd.read_csv('spam.csv')
```

```python
spam['v1'].value_counts()
```

```python
spam.iloc[5]['v2']
```

```python
# 중복 데이터 확인
spam['v2'].nunique()
```

```python
# 중복 데이터 제거
spam.drop_duplicates(subset=['v2'], inplace=True)
len(spam)
```

```python
# 특수 문자 제거 및 소문자 변경
import re

def cleanText(data):
    text = re.sub('[^a-zA-Z]', ' ', data).lower()
    return text

spam['v2'] = spam['v2'].apply(cleanText)
```

## Tokenizing


```python
token_spam = spam.copy()
stop_words_spam = spam.copy()
stem_spam = spam.copy()
```

```python
import nltk
```

```python
# 예시 코드 코드
from nltk.tokenize import word_tokenize, WordPunctTokenizer, TreebankWordTokenizer

nltk.download('punkt')
print(word_tokenize(spam.iloc[5]['v2']))
print(WordPunctTokenizer().tokenize(spam.iloc[5]['v2']))
print(TreebankWordTokenizer().tokenize(spam.iloc[5]['v2']))
```

```python
token_spam['v2'] = spam.apply(lambda row: word_tokenize(row['v2']), axis=1)
```

### 불용어 제거

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stop_words = stopwords.words('english')
```

```python
stop_words_spam['v2'] = token_spam.apply(lambda row: [word for word in row['v2'] if word not in stop_words], axis=1)
```

### 어간 추출

```python
from nltk.stem import PorterStemmer
nltk.download('wordnet')
nltk.download('omw-1.4')

stemmer = PorterStemmer()
print([stemmer.stem(word) for word in stop_words_spam.iloc[5]['v2']])
```

```python
stem_spam['v2'] = stop_words_spam.apply(lambda row: [stemmer.stem(word) for word in row['v2']], axis=1)
```

## Embedding

- 수업에서 다룬 임베딩 방법에는 One-hot encoding, CBOW, Skip-gram, GloVe, FastText가 있었습니다. 다양한 시도와 '비교' 결과를 함께 적어주세요! 파라미터를 조정해가는 과정도 해석에 도움이 될 수 있겠죠 :)

```python
from sklearn.feature_extraction.text import CountVectorizer

ct_vector1 = CountVectorizer(max_features= 500, stop_words = 'english', lowercase= False)
corpus_ham = sum(stem_spam[stem_spam.v1 == 'ham']['v2'], []) 
ct_ham = ct_vector1.fit_transform(corpus_ham)
```

```python
count_ham = pd.DataFrame( {'word' : ct_vector1.get_feature_names(),
                           'count' : ct_ham.sum(axis = 0).flat})

count_ham.sort_values('count', ascending = False).head(3) # 상위 3개
```

```python
ct_vector2 = CountVectorizer(max_features= 500, stop_words = 'english', lowercase= False)
corpus_spam = sum(stem_spam[stem_spam.v1 == 'spam']['v2'], []) 
ct_spam = ct_vector2.fit_transform(corpus_spam)
```

```python
count_spam = pd.DataFrame( {'word' : ct_vector2.get_feature_names(),
                           'count' : ct_spam.sum(axis = 0).flat})

count_spam.sort_values('count', ascending = False).head(3) # 상위 3개
```

```python
from itertools import chain
flatten = chain.from_iterable(stem_spam['v2'].values)
words = list(set(flatten))
print(len(words))
```

### Word2Vec(skip-gram)

```python
from gensim.models import Word2Vec
spam_model = Word2Vec(stem_spam[stem_spam.v1 == 'spam']['v2'], min_count=5, window=2, sg=1) # skip-gram
```

```python
spam_model.wv.most_similar("free")
```

```python
ham_model = Word2Vec(stem_spam[stem_spam.v1 == 'ham']['v2'], min_count=5, window=2, sg=1) # skip-gram
```

```python
ham_model.wv.most_similar("come")
```

### FastText

```python
from gensim.models import FastText

# spam
spam_fast_model = FastText(stem_spam[stem_spam.v1 == 'spam']['v2'], window=5, min_count=5, workers=4, sg=1)
```

```python
spam_fast_model.wv.most_similar("free")
```

```python
ham_fast_model = FastText(stem_spam[stem_spam.v1 == 'ham']['v2'], window=5, min_count=5, workers=4, sg=1)
```

```python
ham_fast_model.wv.most_similar("come")
```
