## 라이브러리 불러오기

```python
import numpy as np
from matplotlib.pylab import rcParams
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf 
from statsmodels.tsa.seasonal import seasonal_decompose 
import matplotlib.pyplot as plt
from pmdarima.arima import ndiffs
import pmdarima as pm
```

## 데이터 불러오기

```python
col_names = ["date", "value"]
df = pd.read_csv('./Electric_Production.csv',names = col_names, header = 0, parse_dates = [0])
df = df.set_index(['date'])
df.head()
```

## 식별

```python
# 12개월의 이동 평균
rolling_mean = df.rolling(window=12).mean()
# 12개월의 이동 분산
rolling_std = df.rolling(window=12).std()
plt.figure(figsize = (10,6))
plt.plot(df, color='cornflowerblue', label='Original')
plt.plot(rolling_mean, color='firebrick', label='Rolling Mean')
plt.plot(rolling_std, color='limegreen', label='Rolling Std')
plt.xlabel('Date', size = 12)
plt.ylabel('Electric Production', size  = 12)
plt.legend(loc = 'upper left')
plt.title('Rolling Statistics', size = 14)
plt.show()
```

#### 위 시계열 데이터는 위로 증가하는 추세를 가지고 있기 때문에 비정상 시계열 데이터입니다.
#### 그렇기 때문에 차분을 통해 비정상 시계열 데이터를 정상 시계열데이터로 변환합니다. 

```python
# 1차 차분 데이터
first_order = df.copy()
# log scale적용
first_order['value'] = np.log1p(first_order['value'].diff())
first_order.dropna(inplace=True)
```

```python
plot_acf(first_order)
plot_pacf(first_order)
plt.show()
```

#### 차분을 1로 하고 log scale로 정규화한 데이터를 ACP, PACP 해보았습니다. 
#### 그 결과 p = 2, d = 1, q = 2인 ARIMA모델이 적합하다고 판단되었습니다. 

## 추정

```python
y = df['value'].values
y_train, t_test = y[1:len(y)-10], y[len(y)-10:]
model = pm.auto_arima(y_train, d=1, start_p=0, max_p=2,start_q =0, max_q=2,seasonal=False, trace = True)
fit = model.fit(y_train)
```
