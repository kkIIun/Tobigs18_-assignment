# 2. 결측치, 이상치 검토

```python
# 결측치는 없음
data.isna().sum()
```

```python
f, axes = plt.subplots(2, 2, figsize=(10,10))
sns.boxplot(y='MonthlyIncome',data=data, ax=axes[0][0])
sns.boxplot(y='YearsAtCompany',data=data, ax=axes[1][0])
sns.boxplot(y='TotalWorkingYears',data=data, ax=axes[0][1])
sns.boxplot(y='YearsWithCurrManager',data=data, ax=axes[1][1])
```

```python
def remove_outlier(d_cp, column):
    fraud_column_data = d_cp[column]
    quan_25 = np.percentile(fraud_column_data, 25)
    quan_75 = np.percentile(fraud_column_data, 75)
    
    iqr = quan_75 - quan_25
    iqr = iqr*1.5
    lowest = quan_25 - iqr
    highest = quan_75 + iqr
    outlier_index = fraud_column_data[(fraud_column_data < lowest) | (fraud_column_data > highest)].index
    print(len(outlier_index))
    d_cp.drop(outlier_index, axis=0, inplace=True)
    print(d_cp.shape)
    return d_cp
```

```python
data = remove_outlier(data, 'MonthlyIncome')
data = remove_outlier(data, 'YearsAtCompany')
data = remove_outlier(data, 'TotalWorkingYears')
data = remove_outlier(data, 'YearsWithCurrManager')
```

###### 총 220개의 outlier들을 제거했습니다.

# 3. 상관관계 파악

```python
plt.figure(figsize=(10,10))
sns.heatmap(data.corr())
```

```python
plt.scatter(data['MonthlyIncome'], data['JobLevel'])
```

##### 월급과 업무의 수준이 양의 상관관계를 갖고 있습니다.

```python
dummy_variable_1 = pd.get_dummies(data["Attrition"])
dummy_variable_1.rename(columns={'No':'Attrition_No', 'Yes':'Attrition_Yes'}, inplace=True)
data = pd.concat([data, dummy_variable_1], axis=1)
data.drop("Attrition", axis = 1, inplace=True)
```

```python
Sales, HR, RD = data.groupby('Department')['Attrition_Yes'].agg(np.mean).sort_values(ascending=False)
```

```python
exp_val = [Sales, HR, RD]
exp_label = ['Sales', 'Human Resources', 'Research & Development']
palette_color = sns.color_palette('bright')
plt.pie(exp_val,labels=exp_label,radius=1.5,colors=palette_color,autopct='%0.2f%%',normalize=True);
```

##### 퇴사율이 가장 높은 부서는 영업입니다.

```python
data.groupby('BusinessTravel')['Age'].agg(len).sort_values(ascending=False)
```

```python
sns.set(rc={'figure.figsize':(18,10)})
plot=sns.catplot(x='BusinessTravel',hue='Department',kind='count',data=data)
```

###### 부서 대부분 직원들은 여행을 드물게 갑니다.

```python
HR, TD, Marketing, LS, Other, Medical = data.groupby('EducationField')['Attrition_Yes'].agg(np.mean).sort_values(ascending=False)
```

```python
exp_val=[HR, TD, Marketing, LS, Other, Medical]
exp_label=['Human Resources','Technical Degree','Marketing','Life Sciences','Medical','Other']
palette_color = sns.color_palette('bright')
plt.pie(exp_val,labels=exp_label,radius=1,colors=palette_color,autopct='%0.2f%%', normalize=True);
```

###### 가장 퇴사율이 높은 전공은 인사관리학입니다.

```python
data.groupby('Age')['Attrition_Yes'].agg(np.mean).sort_values(ascending=False)
```

###### 18~21 나이가 가장 퇴사율이 높습니다.
